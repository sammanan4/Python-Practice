<h1> Adding, deleting rows and columns </h1>

```python
countries
```

```python
# combining two columns
countries['name and code'] = '(' + countries['name'] + ', ' + countries['code'] + ')'
```

```python
countries
```

```python
# deleting columns
countries.drop(columns=['name', 'code'], inplace=True)
```

```python
countries
```

```python
# get the 2 columns back from the name and code column
# the slice removes the paranthesis around the item and then we split the two
# expand converts the list of values into two lists
countries[['name', 'code']] = countries['name and code'].str.slice(start=1, stop=-1).str.split(', ', expand=True).iloc[:, [0,1]]
```

```python
countries
```

```python
# remove the added colimn
countries.drop(columns=['name and code'], inplace=True)
countries
```

```python
# adding and removing rows
# remove all the rows where code starts with A
filt = countries['code'].str.startswith('A')
countries.drop(index=countries[filt].index)
```

```python
# add my own country at the end
# ignore index is necessary when appending
# sort param sorts the columns
countries.append({'name': 'manan'}, ignore_index=True, sort=True)
```

```python
# add my country at a specific index
# here we are slicing the original dataframe and insert the new item
# and then reset the index (drop = False adds a new column named index so keep it True)
pd.concat([countries[:1], pd.DataFrame({'name':'manan'}, index=[1]), countries[1:]]).reset_index(drop=True)
```

```python
# sort based on index
countries.sort_index(ascending=False, kind='mergesort')
```

```python
# sort by rand (descending) and in case of conflicts then sort by name (ascending)
countries.sort_values(['rand', 'name'], ascending=[False, True])
```

```python
# sort by rand (descending) and in case of conflicts then sort by phone (ascending)
# ignore index changes the indexes (see above cell where ignore_index=False by default)
countries.sort_values(['rand', 'phone'], ascending=[False, True], ignore_index=True,)
```

```python
countries.sort_index(axis=1)
```

```python
# using pd concat to create a dataframe from two series having same index columns
codes = countries.set_index('name')['code']
capitals = countries.set_index('name')['capital']
pd.concat([codes, capitals], axis='columns')
```
